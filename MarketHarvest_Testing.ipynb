```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def GIMP_HSV_To_OpenCV_HSV(gimpH, gimpS, gimpV):
    
    opencvH = float(gimpH / 2)
    opencvS = float(gimpS / 100) * 255
    opencvV = float(gimpV / 100) * 255

    return (opencvH, opencvS, opencvV)

def MarketHarvest(file, gimpH, gimpS, gimpV, radius):

    image = cv2.imread(str(file))
    rgb = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    plt.imshow(rgb)

    min_hsv_range=GIMP_HSV_To_OpenCV_HSV(gimpH[0], gimpS[0], gimpV[0])
    max_hsv_range=GIMP_HSV_To_OpenCV_HSV(gimpH[1], gimpS[1], gimpV[1])

    image_bgr = cv2.imread(str(file))
    image = image_bgr.copy()
    image_hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)

    mask = cv2.inRange(image_hsv,min_hsv_range, max_hsv_range)

    #sets the pixels which fall between the color range to white and the rest to black
    plt.imshow(mask)

    cnts,_ = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL,
            cv2.CHAIN_APPROX_SIMPLE)
    c_num=0
    for i,c in enumerate(cnts):
        # draw a circle enclosing the object
        ((x, y), r) = cv2.minEnclosingCircle(c)
        if r>radius:
            c_num+=1
            cv2.circle(image, (int(x), int(y)), int(r), (0, 255, 0), 2)
            cv2.putText(image, "#{}".format(c_num), (int(x) - 10, int(y)), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)
        else:
            continue
    print(c_num)
MarketHarvest('oranges-on-tree.jpeg', (20.4, 49.1), (97.6, 100.0), (54.1, 90.6), 8)
```
