## 1. Upload Images

```python
from google.colab import files
# Prompt the user to upload the file
images = files.upload()

image_paths = []

# Access the uploaded file
for filename, file_content in images.items():
    print(f"Uploaded file: {filename}, size: {len(file_content)} bytes")
    image_paths.append(filename)
```

## 2. Run Inference

```python
!pip install ultralytics
```

```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11m.pt")  # pretrained YOLO11n model

# Run batched inference on a list of images
results = model(image_paths)  # return a list of Results objects
```

## 3. View results

```python
i = 0
for result in results:
    boxes = result.boxes  # Boxes object for bounding box outputs
    result.save(filename=f"result{i}.jpg")  # save to disk
    i += 1
```
